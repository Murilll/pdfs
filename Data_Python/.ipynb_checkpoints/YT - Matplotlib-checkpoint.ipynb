## Importação

```python
import matplotlib.pyplot as plt
```

### plot()
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html?highlight=pyplot%20plot#matplotlib-pyplot-plot

```python
x=[1,2,3,4]
y=[2,3,4,3]
plt.plot(x,y)
plt.show()
```

```python
x=[0,1,2,3,4,5]
y=[0,1,2,3,4,5]

plt.plot(x,y,label="Dados")
plt.title("Título")
plt.legend()

plt.ylabel("Eixo Y")
plt.xlabel("Eixo X")

plt.xticks([0,1,2,3,4,5,6])
plt.yticks([0,1,2,3,4,5,6])


plt.show()
```

## LINES propriedades
https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_marker

```python
plt.plot(x,y,linestyle="dashed") #linestyle: Mudar o Estilo de Linha OU udar ls as linestyle
plt.show()
```

```python
plt.plot(x,y,ls="solid",color="red") #color: P/ Mudar cor da linha
plt.show()
```

```python
plt.plot(x,y,ls="dashdot",color="black",lw=4) #lw: P/ Mudar espessura da linha 
plt.show()
```

## .Scatter()

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib-pyplot-scatter

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.scatter(x,y,color="black")
plt.show()
```

## .bar()

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib-pyplot-bar

```python
plt.bar(x,y,color="green")
plt.show()
```

### Nomeando os eixos de um gráfico:

### Mudando os tipos de marcadores do gráfico :
https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers

```python
plt.scatter(x,y,color="green",marker="s")
plt.show()
```

### Alterando os limites dos gráficos:
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axis.html#matplotlib.pyplot.axis

```python
plt.plot(x,y)
plt.axis(xmin=0,xmax=100,ymin=0,ymax=100) # Colocar os parâmetros do eixo x e y
plt.show()
```

```python
plt.plot(x,y)
plt.axis("auto") # Marca os parâmetros automaticamente 
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
x1=[2,4,6,8,10]

plt.plot(x,y,x1,y)
plt.axis("square") # Mostra o gráfico quadriculado de forma "auto"
plt.show()
```

### FIGURE
https://matplotlib.org/stable/api/figure_api.html?highlight=sharex#module-matplotlib.figure

### SUBPLOTS
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html


```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]


figura= plt.figure(figsize=(20,5))
figura.suptitle("Título Geral")


figura.add_subplot(131)  # (linha,coluna,posição)
plt.plot(x,y,label="Dados")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.title("Gráfico 1")

figura.add_subplot(132)  #(Na linha 1, da coluna 3, posição 2)
plt.bar(x,y,label="Dados")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.title("Gráfico 2")

figura.add_subplot(133)
plt.scatter(x,y,label="Dados")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.title("Gráfico 3")


#plt.savefig("Nome.png") P/ Salvar a Figura

plt.show()
```
